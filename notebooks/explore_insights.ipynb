```python
#import libraries & Load Cleaned Data

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Setup
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)

# Load cleaned data
df = pd.read_csv("../data/processed/flipkart_cleaned.csv")
df.head()
```

```python
# Top 10 Discounted products

df["discount_percent"] = ((df["retail_price"] - df["discounted_price"]) / df["retail_price"]) * 100
top_discounts = df.sort_values(by="discount_percent", ascending=False).head(10)

top_discounts[["product_name", "brand", "retail_price", "discounted_price", "discount_percent"]]
```

```python
# Top brands by number of products.

top_brands = df["brand"].value_counts().head(10)

sns.barplot(x=top_brands.values, y=top_brands.index, palette="viridis")
plt.title("Top 10 Brands by Product Count")
plt.xlabel("Number of Products")
plt.ylabel("Brand")
plt.show()
```

```python
# Average Discount % per Brand

avg_discount_by_brand = df.groupby("brand")["discount_percent"].mean().sort_values(ascending=False).head(10)

sns.barplot(x=avg_discount_by_brand.values, y=avg_discount_by_brand.index, palette="mako")
plt.title("Top 10 Brands by Average Discount %")
plt.xlabel("Average Discount %")
plt.ylabel("Brand")
plt.show()
```

```python
# price distribution

# Cap the price at ₹10,000 for better visibility
price_cap = 10000
filtered_df = df[df["discounted_price"] <= price_cap]

sns.histplot(filtered_df["discounted_price"], bins=40, kde=True, color="steelblue")
plt.title(f"Distribution of Discounted Prices (Up to ₹{price_cap})")
plt.xlabel("Discounted Price (₹)")
plt.ylabel("Product Count")
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
#Top 10 categories by product count.

# Extract main category from 'product_category_tree'
# Convert from string list format and extract the first category
df['main_category'] = df['product_category_tree'].str.extract(r"\[\"(.*?)\"\]")  # Remove brackets
df['main_category'] = df['main_category'].str.split(">>").str[0].str.strip()    # Get first category

# Top 10 categories
top_categories = df['main_category'].value_counts().head(10)

# Plot
sns.barplot(x=top_categories.values, y=top_categories.index, palette="viridis")
plt.title("Top 10 Product Categories")
plt.xlabel("Number of Products")
plt.ylabel("Category")
plt.show()
```

```python
# Step 1: Only extract if the field looks like a valid category list
valid_category_rows = df['product_category_tree'].str.startswith('["')

# Step 2: Apply extraction only to valid rows
df.loc[valid_category_rows, 'main_category'] = (
    df.loc[valid_category_rows, 'product_category_tree']
    .str.extract(r'\["(.*?)"\]')[0]  # Extract inner string
    .str.split(">>").str[0].str.strip()
)

# Step 3: Drop rows with missing or overly long main categories
df = df[df['main_category'].notna() & (df['main_category'].str.len() < 30)]


# Calculate discount amount if not already

df['discount_amount'] = df['retail_price'] - df['discounted_price']

# Drop rows where main_category is missing or corrupted
df_clean = df.dropna(subset=['main_category', 'discount_amount'])

#print("🔍 Sample values before grouping:\n", df[['main_category', 'discount_amount']].head(10))


# Group by correct main_category
avg_discount = (
    df.groupby('main_category', as_index=True)['discount_amount']
    .mean()
    .sort_values(ascending=False)
    .head(10)
)

# View top entries
#print("🔍 Top Average Discounts by Category:\n", avg_discount)

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(x=avg_discount.values, y=avg_discount.index, palette='coolwarm')
plt.title("Average Discount Amount by Main Category")
plt.xlabel("Average Discount (INR)")
plt.ylabel("Main Category")
plt.tight_layout()
plt.show()
```

```python
df.to_csv("flipkart_cleaned.csv", index=False)
```

```python
import numpy as np

# Replace "No rating available" with NaN
df['product_rating'] = df['product_rating'].replace("No rating available", np.nan)
df['overall_rating'] = df['overall_rating'].replace("No rating available", np.nan)

# Combine the two columns into one unified 'rating' column
# Priority: product_rating > overall_rating
df['rating'] = df['product_rating'].combine_first(df['overall_rating'])

# Convert rating column to float
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

# Drop the original columns if not needed
df.drop(['product_rating', 'overall_rating'], axis=1, inplace=True)

# Check how many ratings are available now
print("Number of products with valid ratings:", df['rating'].notna().sum())
```

```python
# Save the cleaned dataset, overwriting the existing file
df.to_csv("../data/processed/flipkart_cleaned.csv", index=False)
print("Cleaned dataset saved successfully.")
```

```python
# Discount% vs Discounted Price

plt.figure(figsize=(10, 6))
sns.regplot(
    data=df,
    x='discounted_price',
    y='discount_percent',
    scatter_kws={'alpha': 0.3},
    line_kws={"color": "red"}
)
plt.title('Discount % vs Discounted Price')
plt.xlim(0, 3000)  # or try 3000, depending on spread
plt.ylim(0, 100) 
plt.xlabel('Discounted Price (₹)')
plt.ylabel('Discount Percentage (%)')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

🧠 A few quick observations you might note in above analysis:
1. Heavily discounted products (above 60–80%) are more frequent at lower price ranges (₹0–₹1000).
2. The regression line is almost flat or slightly declining — indicating:
    On average, higher-priced products tend to have slightly lower discount percentages.

```python
# Relationship between Ratings and prices.

import numpy as np

# Convert ratings to numeric
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

# Get indices where rating is NaN
nan_indices = df[df['rating'].isna()].index

# Fill NaNs with random values between 1.0 and 5.0 (1 decimal precision)
random_ratings = np.round(np.random.uniform(1.0, 5.0, size=len(nan_indices)), 1)

# Fill NaNs with those random values
df.loc[nan_indices, 'rating'] = random_ratings

df.to_csv("../data/processed/flipkart_cleaned.csv", index=False)
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='discounted_price', y='rating', alpha=0.4, color='purple')
sns.regplot(data=df, x='discounted_price', y='rating', scatter=False, color='red')

plt.title('Rating vs Discounted Price (With Imputed Ratings)')
plt.xlim(0, 50000) 
plt.xlabel('Discounted Price (₹)')
plt.ylabel('Rating')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

 Interpretation: Rating vs Discounted Price
🟥 What the plot shows:
1. A slightly negative regression line (red) means:
    👉 Higher-priced products tend to have lower or similar ratings, not necessarily better.

2. The ratings are clustered around 3.0–5.0 for all price ranges.

3. A few extremely high-priced outliers (₹100K–₹500K+) distort the plot and spread the regression band.

📉 Insight:

There's no strong correlation between product price and rating.
Expensive products don’t guarantee better customer satisfaction on Flipkart, based on available (and imputed) rating data.
